```python
def tmp_agg(i):
    return np.mean([j >= 0 for j in i if j != 0])
```

```python
def load_qrel_approach(csv):
    df = pd.read_csv('/mnt/ceph/storage/data-in-progress/data-research/web-search/ir-axioms/preferences/2022-02-18/' + csv)
    df['run_type'] = df['name'].apply(lambda i: TYPES[csv][i])
    already_covered = set()
    
    df_analysis = []
    for _, doc_pair in tqdm(df.iterrows(), total=len(df)):
        ids = str(doc_pair['docno_a']) + '__' + str(doc_pair['docno_b'])
        if doc_pair['ORACLE_preference'] > 0:
            if ids not in already_covered:
                row = {}
                for axiom, axiom_name in AXIOM_DISPLAY_NAMES.items():
                    row[AXIOMATIC_GROUPS[axiom_name]] = doc_pair[axiom]
                df_analysis += [row]
        already_covered.add(ids)
    return pd.DataFrame(df_analysis)
```
